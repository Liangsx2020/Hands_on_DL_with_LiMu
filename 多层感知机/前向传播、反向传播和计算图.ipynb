# 前向传播、反向传播和计算图



## 前向传播

前向传播指的是：按顺序（从输入层到输出层）计算和存储神经网络中每层的结果。

研究单隐藏层神经网络的机制，假设输入样本是$\mathbf{x} \in \mathbb{R}^{d}$，并且隐藏层中不包含偏置项。这里的中间变量是:
$$
\mathbf{z} = \mathbf{W}^{(1)} \mathbf{x},
$$
其中，$\mathbf{W}^{(1)} \in \mathbb{R}^{h \times d}$是隐藏层的权重参数。将中间变量$\mathbf{z} \in \mathbb{R}^{h}$送入激活函数$\phi$之后，得到长度为$h$的隐藏激活向量:
$$
\mathbf{h} = \phi(\mathbf{z}).|
$$
隐藏变量$\mathbf{h}$也是一个中间变量。假设输出层的参数只有权重$\mathbf{W}^{(2)} \in \mathbb{R}^{q \times h}$，我们可以得到输出层变量，他是一个长度为$q$的向量：
$$
\mathbf{\omicron} = \mathbf{W}^{(2)} \mathbf{h}.
$$
假设损失函数是$l$，样本标签为$y$，我们可以计算单个数据样本的损失项,
$$
L = l(\mathbf{\omicron}, y).
$$
根据$L_2$正则化的定义，给定超参数$\lambda$, 正则化项是：
$$
s = \frac{\lambda}{2} \left(\|\mathbf{W}^{(1)}\|_F^2 + \|\mathbf{W}^{(2)}\|_F^2\right),
$$
其中矩阵的$F$范数表示矩阵元素的平方和的开方。最后，模型在给定的数据样本上的正则化损失为：
$$
J = L + s.
$$

在下面的讨论中，我们将$J$称为目标函数。

## 反向传播
反向传播指的是计算神经网络中各个参数的梯度，。根据链式求导法则，按相反的顺序从输出层到输入层遍历网络。储存了每个参数的梯度，用于更新参数。

以上一个前向传播为例子。
第一步，计算$J = L + s$相对于损失项$L$和正则项$s$的梯度。
$$
\frac{\partial J}{\partial L} = 1 and \frac{\partial J}{\partial s} = 1
$$
接下来根据链式法则计算目标函数关于输出层变量$\omicron$的梯度。
$$
\frac{\partial J}{\partial \omicron} = \frac{\partial J}{\partial L} \cdot \frac{\partial L}{\partial \omicron} = \frac{\partial L}{\partial \omicron}
$$
计算正则化参数相对于两个参数的梯度。
$$
\frac{\partial s}{\partial \mathbf{W}^{(1)}} = \lambda \mathbf{W}^{(1)} and \frac{\partial s}{\partial \mathbf{W}^{(2)}} = \lambda \mathbf{W}^{(2)}
$$
